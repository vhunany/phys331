```python
import numpy as np
import matplotlib.pyplot as plt

def plotfunc(a, xInput):
    """
    Returns the y-values for function tanh(ax) over the domain âˆ’5 â‰¤ x â‰¤ 5. 
    
    Accepts a parameter `a` to scale the input `x`. `a` is some real number constant.
    """

    yValues = np.tanh(a * xInput)
    
    return(yValues)
    
    

def main_bc():
    """
    Entry point to my program. 
    
    Generates a plot of the tanh(ax) function for a = 0.5, 1.3, and 2.2 on the same graph.
    
    Uses the plotfunc function to create the plots, labels the x and y axes, 
    and adds a legend. The plot is displayed at the end of the main_bc function.
    """
        
    # Defining limits and mesh spacing
    x1 = -5.
    x2 = 5.
    dx = 0.1

    a_list = [0.5, 1.3, 2.2]  # list of a values to iterate with 

    for a_num in a_list:  # iterating through our list of a's 

        # Creating a function for each function with different a values 
        
        # Defining x axis and y values
        xAxis = np.arange(x1,x2,dx)
        #yValues = np.tanh(xAxis)
        yValues = plotfunc(a_num, xAxis)
        
        # generating plot
        plt.plot(xAxis, yValues, label=f'a = {a_num}')  # adding the plot to the existing plot object
    
    plt.xlabel("x values where âˆ’5 â‰¤ x â‰¤ 5")
    plt.ylabel("y values for tanh(ax)")
    plt.title("Graphs of hyperbolic function, tanh(ax), graphed from domain of âˆ’5 â‰¤ x â‰¤ 5 where a is some constant")
    plt.legend()
    plt.show()

main_bc()
```
