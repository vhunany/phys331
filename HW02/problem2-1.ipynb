```python
import numpy as np

def calculate_pi(N):
    x = np.random.uniform(-1, 1, N) # should we include 1? 
    y = np.random.uniform(-1, 1, N)

    # x and y create a square. Now we have to check if a point lies inside of the circle using
    # the equation x^2 + y^2 = r^2 
    r = 1 
    i = 1 
    counter = 0 

    while i < N:  # Is it ok that I have it as a while loop instead of using np.array(x[i])**2 + (y[i])**2 <= r**2)?? 
        if (x[i])**2 + (y[i])**2 <= r**2: 
            counter += 1
        i+=1 
    
    pi = 4 * counter/N
    
    return pi
```

Cell for *Problem 2(b)* is below.

```python
import numpy as np
from matplotlib import pyplot as plt

# Total number of runs
total_runs = 25
actual = np.pi

# Use np.logspace to generate N values logarithmically spaced between 10^2 and 10^6
N_values = np.logspace(2, 6, total_runs, dtype=int)

errors = []

for N_times in N_values:  # iterate through each power of 10 in order to get
    # values from 100 to 1000000 to use in our call to the function 
    measured = calculate_pi(N_times)
    print(f"Measured π for N={N_times}: {measured}")
    
    # Calculate the relative error
    rel_error = np.abs((measured - actual) / actual)
    
    # Store the relative error
    errors.append(rel_error)

# Plot the results on a log-log plot
plt.loglog(N_values, errors, marker='o')
plt.xlabel('Number of Random Points (N)')
plt.ylabel('Relative Error')
plt.title('Relative Error vs. Number of Random Points in π Approximation')
plt.grid(True, which="both", ls="--")
plt.show()
```
