```python
def random_number_generator(a,c,seed,m,N):
    num_list = []

    I = seed 

    for x in range(0, N): 
        I = (a * I + c) % m
        num_list.append(I / m)
    return num_list
```

**Question.**
By trial and error, how high do you have to set to set m (in powers of 2), until the numbers appear truly random? Why is m the relevant variable here?

Through trial and error, I experimented with different values of `m` for generating random numbers. Initially, I set `m` for the x array to powers of 2, starting with \(2^{20}\) and even going as high as \(2^{500}\). Surprisingly, when I increased `m` for both the x and y arrays to very large values (e.g., \(2^{5000}\)), the resulting graph appeared "more random" than before. This suggests that the larger the value of `m`, the more randomness is introduced.

This behavior makes sense. If `m` is small, such as 1, the random number generator can only produce a very limited set of values (in this case, always 1), which results in minimal variation and low randomness. As we increase `m`, the range of possible numbers grows, allowing the random number generator to produce a more diverse set of values. This increased diversity translates to greater randomness in the output.

In conclusion, the larger the modulus `m`, the larger the range of numbers the random number generator can produce, leading to a more random distribution of values.
